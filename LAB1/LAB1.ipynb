```python
import pandas as pd
import seaborn as sn
```

```python
dt = pd.read_csv('data/top_spotify_tracks.csv', header=0)
```

```python
dt
```

```python
dt.shape
```

```python
dt.dtypes
```

```python
dt.describe()
```

```python
dt['id'] = dt['id'].astype('category')
dt['name'] = dt['name'].astype('category')
dt['artists'] = dt['artists'].astype('category')
```

```python
dt['id'] = dt['id'].cat.codes
dt['name'] = dt['name'].cat.codes
dt['artists'] = dt['artists'].cat.codes
```

```python
dt.describe()
```
